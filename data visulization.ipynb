```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df= pd.read_excel("flight.xlsx")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# make column to low column 
df.columns = [str(name).lower() for name in df.columns.tolist()]
df.head()
```

```python
airline_counts = df['airline'].value_counts()

# Generate colors
colors = plt.cm.tab10.colors 

# Plot the data
plt.barh(airline_counts.index, airline_counts.values, color=colors[:len(airline_counts)])
plt.xlabel('Number of counts')
plt.ylabel('Airline')
plt.title('Count of Airlines')
plt.show()
```

```python
top_airline=df['airline'].value_counts()
top_airline = top_airline.nlargest(10)
print(top_airline)
```

```python
top_airline=df['airline'].value_counts()
top_airline = top_airline.nlargest(10)
print(top_airline)

ax =top_airline.plot(kind="bar")
plt.title("Top Airlines")
plt.xlabel("Airlines")
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
airline_codes = df.sort_values(by='airline_code')
airline_codes
```

```python
avg_DEP_delays_Airlines=df.groupby("airline").mean("flight delay").sort_values
avg_DEP_delays_Airlines
```

```python
avg_DEP_delays_Airlines = df.groupby("airline")["dep_delay"].mean().sort_values()
avg_DEP_delays_Airlines
```

```python
plt.bar(avg_DEP_delays_Airlines.index, avg_DEP_delays_Airlines,color='purple')
plt.xlabel('Airline')
plt.ylabel('Mean ARR_delay')
plt.xticks(rotation=90)
plt.show()
```

```python
Airline = df['airline'].value_counts()
plt.figure(figsize=(10, 7))
plt.pie(Airline, labels=Airline.index, autopct='%1.1f%%', startangle=10)
plt.title('Airlines by Share of Domestic Flight Volume')
plt.show()
```

```python
top_origin_counts = airline_counts = df['origin'].value_counts().head(30)
top_origin_counts
```

```python
sns.barplot(x=top_origin_counts.index, y=top_origin_counts.values,color='purple')
plt.xlabel('Origin Airport Name')
plt.ylabel('Count')
plt.title('Top Airports by Outgoing Flight Count')
plt.xticks(rotation=80)
plt.show()
```

```python
dest_city_counts = airline_counts = df['dest_city'].value_counts().head(30)
dest_city_counts
```

```python
dest_city_counts = airline_counts = df['dest_city'].value_counts().head(30)
top_origin_counts = airline_counts = df['origin'].value_counts().head(30)
combined_city_count= dest_city_counts.add(top_origin_counts,fill_value=0)
sorted_combine_count = combined_city_count.sort_values(ascending=False)
top_cities = sorted_combine_count.head(40)
others_count = sorted_combine_count[40:].sum()
top_cities['Others'] = others_count

plt.figure(figsize=(10, 7))
plt.pie(top_cities, labels=top_cities.index, autopct='%1.1f%%', startangle=90)
plt.title('Cities by Share of Flight Count')
plt.show()

```
